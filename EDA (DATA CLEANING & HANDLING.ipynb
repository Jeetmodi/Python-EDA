```python
#Checking values in the Home Ownership column.Now as per our problem statement We can define that Fully paid & Current can 

df.loan_status.value_counts()
```

#### Now as per our problem statement We can define that Fully paid & Current status are non defaulters and Charged off can be define as Defaulters So we are assigning Defaulters = 1 else 0

```python
#As analyized in the columns we can catagorize Loan Amount,Funded Amount,Intrest Rate,Annual Income


df['Loan_cat'] = pd.cut(df.loan_amnt , [0,5600,10000,15000,999999], labels = ["Low","Medium","High","Very High"])

df['fund_cat'] = pd.cut(df.funded_amnt_inv , [0,5000,9000,15000,999999], labels = ["Low","Medium","High","Very High"])

df['int_rate_cat'] = pd.cut(df.int_rate , [0,9.32,11.86,14.65,999999], labels = ["Low","Medium","High","Very High"])

df['annual_cat'] = pd.cut(df.annual_inc , [0,42000,60000,83000,999999], labels = ["Low","Medium","High","Very High"])

df.head()
```
