## Bivariate Analysis

```python
#Creating a fucntion to create the Barchart of Loan Status with other Variables
def barchart(var):
    sns.barplot(x = var,y = 'loan_status' , data = df , ci =None )
    plt.show()   
```

```python
def barplot(var):
    df.groupby(var)['loan_status'].mean().plot.bar(width = 0.8, color = ['slateblue' , 'limegreen' , 'dimgray' , 'tomato' , 'dodgerblue'])
    plt.xticks(rotation = 0)
    plt.show()
```

```python
df.groupby('term')['loan_status'].mean()
```

### On an Average 22% of the Customers who have taken as 60 monthsTerm are analyzed as Defaulters

```python
plt.title('Term Vs. Loan Default')
barchart(df.term)
```

```python
df.groupby('grade')['loan_status'].mean()
```

### The highest defaulters lies in G grade loan at around 32% followed by F and E at 30% and 25% respectively 

```python
plt.figure(figsize=[10,5])
plt.title('Loan Grade Vs. Loan Default')
barplot(df.grade)
```
